## Python Environments

An **environment** is a self-contained collection of Python packages and dependencies. It includes:
- A specific Python interpreter (kernel) version
- Installed packages and their versions
- Environment variables and configuration

Environments ensure your code runs consistently by isolating dependencies from other projects. 

Diligently avoiding global dependencies and maintaining contained project-specific environments will add in circumventing 
- Incomplete records of dependencies for your own projects.
- Difficulty recreating others' local environments.
- Conflicts between projects depending on different versions of the same package.

## UV

We are using uv to manage our environment in this repo. Other options include conda, pipenv, and pyenv.

When you start a Jupyter notebook in VS Code, you select a kernel that's connected to a specific environment. The kernel uses that environment's Python interpreter and packages to execute your code.

In VS Code, you can switch kernels (and thus environments) using the kernel picker in the top-right corner of your notebook, allowing you to run the same notebook with different Python setups as needed.

## Shell Commands（Shell 命令）


Sometimes in a notebook, we want to do things that are beyond the reach of a Python runtime.（有时在笔记本中，我们想做一些超出 Python 运行时能力范围的事情。）

In that case, we can use a special `!` syntax at the beginning of a line to bypass the Python interpreter, and send shell commands to the environment where Python is running.（在这种情况下，我们可以在行首使用特殊的 `!` 语法，绕过 Python 解释器，将 shell 命令发送到 Python 运行所在的环境。）

Using a `!` shell command is roughly equivalent to running that same command in your terminal, though there are some differences we'll discuss soon.（使用 `!` shell 命令大致相当于在终端中运行相同命令，虽然我们很快会讨论其中的一些差异。）

## Environment Secrets（环境机密）


Many exercises in the course will require you to use secret passwords and API keys.（本课程的许多练习将要求你使用机密密码和 API 密钥。）

- Regardless of environment, secrets and keys should never be stored in notebooks or checked into version control.（无论在哪种环境中，都不应将机密信息或密钥存储在笔记本中或提交到版本控制中。）
- Storing and accessing secrets will differ based on whether you are working in a local environment (e.g., on your computer or a remote AWS machine), or in a hosted environment (e.g., Kaggle or Google Colab).（根据你是在本地环境（例如在你的计算机或远程 AWS 机器上）工作，还是在托管环境（例如 Kaggle 或 Google Colab）中工作，存储和访问机密的方式会有所不同。）

In a local environment, with access to the working directory, keys should be:（在本地环境中，如果可以访问工作目录，则应按照以下方式处理密钥：）

- Loaded programatically by a tool like the `python-dotenv` package（使用类似 `python-dotenv` 的工具以编程方式加载）
- Stored in a `.env` file somewhere that can be loaded by python's `load_dotenv` module（存储在可以由 python 的 `load_dotenv` 模块加载的 `.env` 文件中）
- Ignored by version control, if the `.env` file is stored somewhere that version control tracks（如果 `.env` 文件存储在版本控制跟踪的目录中，则应将其加入忽略列表）

This technique assumes that your python project files will be stored in some descendent folder of your home directory.（该方法假定你的 Python 项目文件将存储在主目录的某个子文件夹中。）
That's because the script containing the call to `load_dotenv()` dictates the first folder where a `.env` file will be looked for.（这是因为包含 `load_dotenv()` 调用的脚本决定了查找 `.env` 文件的第一个文件夹。）
After that, `python-dotenv` continues up the chain of parent folders looking for more `.env` files to load from.（之后，`python-dotenv` 会继续向上查找父文件夹，寻找更多要加载的 `.env` 文件。）

Let's work with that module:（让我们使用该模块：）

- [ ] Read & run the following code（阅读并运行以下代码）


```python
# Import the loader and the os namespace where the env variables will be accessible
from dotenv import load_dotenv
import os
```

Now we can try to load a secret we know doesn't exist called `MY_SECRET`, by running:（现在我们可以尝试加载一个我们知道不存在的名为 `MY_SECRET` 的秘密，运行：）


```python
# Load all variables from the .env file 
# so they are available from the os namespace
# 加载所有的变量 .env文件，以便从操作系统名称空间中可用
load_dotenv()

MY_SECRET = os.getenv("MY_SECRET")
print("Loaded from .env file: ", MY_SECRET)
```

Since you haven't defined a .env file yet, the value of `MY_SECRET` in the output above should be `None`.（由于你尚未定义 .env 文件，上述输出中的 `MY_SECRET` 值应该是 `None`。）
