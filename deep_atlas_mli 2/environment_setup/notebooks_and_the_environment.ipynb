## Python Environments

An **environment** is a self-contained collection of Python packages and dependencies. It includes:
- A specific Python interpreter (kernel) version
- Installed packages and their versions
- Environment variables and configuration

Environments ensure your code runs consistently by isolating dependencies from other projects. 

Diligently avoiding global dependencies and maintaining contained project-specific environments will add in circumventing 
- Incomplete records of dependencies for your own projects.
- Difficulty recreating others' local environments.
- Conflicts between projects depending on different versions of the same package.

## UV

## Environment Secrets

Many exercises in the course will require you to use secret passwords and API keys.

- Regardless of environment, secrets and keys should never be stored in notebooks or checked into version control
- Storing and accessing secrets will differ based on whether you are working in a local environment (e.g., on your computer or a remote AWS machine), or in a hosted environment (e.g., Kaggle or Google Colab).

In a local environment, with access to the working directory, keys should be:

- Loaded programatically by a tool like the `python-dotenv` package
- Stored in a `.env` file somewhere that can be loaded by python's `load_dotenv` module
- Ignored by version control, if the `.env` file is stored somewhere that version control tracks

This technique assumes that your python project files will be stored in some descendent folder of your home directory. That's because the script containing the call to `load_dotenv()` dictates the first folder where a `.env` file will be looked for. After that, `python-dotenv` continues up the chain of parent folders looking for more `.env` files to load from.

Let's work with that module:

- [ ] Read & run the following code

```python
# Import the loader and the os namespace where the env variables will be accessible
from dotenv import load_dotenv
import os
```

Now we can try to load a secret we know doesn't exist called `MY_SECRET`, by running:

```python
# Load all variables from the .env file so they are available from the os namespace
load_dotenv()

MY_SECRET = os.getenv("MY_SECRET")
print("Loaded from .env file: ", MY_SECRET)
```

Since you haven't defined a .env file yet, the value of `MY_SECRET` in the output above should be `None`.

Let's create your .env file and add a secret by running the following command.

- [ ] Review and run the code cells below

```python
# Let's set up your .env file the the root coursework folder (deep_atlas_mli/.env)
!echo 'MY_SECRET="abc123"' > ../.env

# (Re)load all variables from the .env file so they are available from the os namespace
load_dotenv()

MY_SECRET = os.getenv("MY_SECRET")
print("Loaded from .env file: ", MY_SECRET)
```
