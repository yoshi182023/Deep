## Kernels
## 内核

Even though you're able to read and interact with this file in the editor you're using, you won't be able to run any code cells until you select an execution engine to compile and interpret the Python. This execution engine is called a "Jupyter kernel".
即使你能够在当前使用的编辑器中读取并与此文件交互，在你选择一个执行引擎来编译和解释 Python 之前，你仍然无法运行任何代码单元。这个执行引擎被称为“Jupyter 内核”。

The kernel has a number of responsibilities:
内核有多个职责：

- Wrap a particular Python executable.
- 封装特定的 Python 可执行文件。
- Launch a process from that executable when the notebook starts, and whenever the user requests a restart.
- 在笔记本启动时以及用户请求重启时，从该可执行文件启动一个进程。
- Manage the interface the Python process has to the operating system environment that process runs in.
- 管理 Python 进程与其运行的操作系统环境之间的接口。
- Broker code execution requests to that process, as well as the output from it.
- 将代码执行请求及其输出中介给该进程。
- Provide editor-related features that go beyond what a simple Python process offers (tab completion, introspection, etc).
- 提供超出简单 Python 进程所能提供的编辑器相关功能（如制表符补全、内部查看等）。

Most of those characteristics are implicitly configured to usable defaults, but you actually want to be careful which kernel you run a notebook on.
这些特性大多数隐式配置为可用的默认值，但你实际上需要注意运行笔记本时选择哪个内核。

- In an ideal world, you'd always run a notebook in a totally bespoke environment, totally separate from your user environment.
- 在理想情况下，你应该始终在一个完全定制的环境中运行笔记本，与用户环境完全隔离。
- In practice, unless you take deliberate steps to prevent it, kernels wind up inheriting certain runtime characteristics from the environment they're running on.
- 实际上，除非你采取刻意措施，否则内核最终会继承其运行环境中的某些运行时特性。

## Cell Outputs
## 单元格输出

One of the big ways notebooks support the "field journal" style of documenting big data projects is by saving code cell output in the notebook file itself.
笔记本支持“现场日志”式记录大数据项目的一大方式是将代码单元格输出保存在笔记本文件本身中。

- [ ] Note how the output is displayed under the code cell above. 
- [ ] 注意输出如何显示在上方代码单元格下方。


In addition to the output below the cell, you'll also see a count of how many times a code cell has been run since last cleared. Furthermore, all that output is saved alongside the rest of the contents of the file. This includes any output images generated by graphing and plotting modules. After running a code cell, you may not want to save the output in the file, particularly if you plan to check the file in to version control. On the other hand, this metadata may also be useful to future-you or other readers, so the choice to keep or discard cell output is subjective.
除了单元格下方的输出外，你还会看到自上次清除以来代码单元格运行了多少次的计数。此外，所有这些输出都会与文件的其余内容一起保存。这包括由绘图模块生成的任何输出图像。运行代码单元格后，你可能不想将输出保存到文件中，尤其是如果你打算将文件提交到版本控制中。另一方面，这些元数据对未来的你或其他读者也可能有用，因此是否保留或丢弃单元格输出是主观的选择。

In any notebook editor, you'll find options to:
在任何笔记本编辑器中，你都会找到以下选项：

- Clear an individual cell output and counter from the `...` options next to the cell.
- 从单元格旁边的 `...` 选项中清除单个单元格输出和计数器。
- Clear all cell outputs and counters in the whole notebook.
- 清除整个笔记本中的所有单元格输出和计数器。

## Running Many Cells All at Once

有时，在多次运行笔记本中的各种代码单元后，或者以不同顺序运行后，你会决定通过从顶部运行所有代码来重新初始化变量。如果你不想点击很多代码单元来重新初始化机器状态，你可以运行给定点之前或之后的所有单元格。
